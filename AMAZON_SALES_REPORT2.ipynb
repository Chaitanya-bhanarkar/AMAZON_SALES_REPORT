### 4.Fulfilment

```python
fulfilment = A_data['Fulfilment'].value_counts()
fulfilment
```

```python
ax = sns.countplot(data=A_data,x='Fulfilment')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Distribution of Fulfilment');
```

### the most of the order fulfill by AMAZON

```python
plt.pie(fulfilment, labels= fulfilment.index, autopct = '%1.1f%%',startangle=90)
plt.title('Distribution of Fulfilment column')
plt.show()
```

```python
fig, cls = plt.subplots()
cls.pie(fulfilment, labels = fulfilment.index, autopct='%1.1f%%', radius=1.5, wedgeprops = dict(width=1))
cls.set(aspect = 'equal')

plt.show()
```
